```python
import pandas as pd
```

```python
df_daten = pd.read_json("dump-stg-api.json", orient="records")
print(df_daten)
```

```python
df_daten_clean = df_daten
```

```python
df_na_rows = df_daten_clean.isna()
```

```python
df_daten_clean
```

```python
df_daten_clean["_id"][0]["$oid"]
```

```python
for i in range(len(df_daten_clean)):
    df_daten_clean["_id"][i] = df_daten_clean["_id"][i]["$oid"]
```

```python
df_structure = pd.DataFrame(columns=["id", "correct", "query", "date", "task"])
```

```python
df_structure
```

```python
index_value = 0
for i in range(len(df_daten_clean)):
    print(i)
    if df_na_rows["history"][i]:
        pass
    else:
        split = df_daten_clean["history"][i]
        print(split)
        keys = split.keys()
        for n in keys:
            for k in range(len(split[n])):
                index_value += 1
                dict_copy = split[n][k]
                new_line = pd.DataFrame({"correct": dict_copy["correct"], "query": dict_copy["query"] , "date": dict_copy["date"], "task": n, "id": df_daten_clean["_id"][i]}, index=[index_value])
                df_structure = df_structure.append(new_line)
```

```python
df_structure
```

```python
pd.DataFrame.to_csv(df_structure)
```
